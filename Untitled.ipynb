```python
from comp5318_assignment1 import models, model_tools, decomposition
import operator as op
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
mr = model_tools.ModelRunner()
mr.load_data()
model = models.GNB()
```

```python
model.fit(mr.xtr, mr.ytr)
preds = model.predict(mr.xte)
```

```python
plt.figure(figsize=(10, 10))
model_tools.plot_confusion_matrix(model_tools.confusion_matrix(preds, mr.yte), ['T-shirt/top','Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'])
plt.title(f'Confusion matrix of GNB model\nAccuracy = {np.mean(preds==mr.yte)}')
plt.show()
```

```python
model = models.KNN(3)
model.fit(mr.xtr, mr.ytr)
preds = model.predict(mr.xte)
```

```python
plt.figure(figsize=(10, 10))
model_tools.plot_confusion_matrix(model_tools.confusion_matrix(preds, mr.yte), ['T-shirt/top','Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'])
plt.title(f'Confusion matrix of KNN model (k=3)\nAccuracy = {np.mean(preds==mr.yte)}')
plt.show()
```
